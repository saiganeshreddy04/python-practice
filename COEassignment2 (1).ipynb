```python
# Total yearly sales for all stores
sales_data[['Jan', 'Feb', 'Mar', 'Apr']].sum().sum()
```

```python
# Total salesm for each month
sales_data[sales_data['city'] == 'California'][['Jan', 'Feb', 'Mar', 'Apr']].sum()
sales_data[sales_data['city'] == 'Texas'][['Jan', 'Feb', 'Mar', 'Apr']].sum()
sales_data[sales_data['city'] == 'Arizona'][['Jan', 'Feb', 'Mar', 'Apr']].sum()
```

```python
#average monthly sales per city
sales_data[sales_data['city'] == 'California'][['Jan', 'Feb', 'Mar', 'Apr']].mean()
```

```python
#city with the highest average sales
cities = sales_data['city'].unique()
city_avg = {city: sales_data[sales_data['city'] == city][['Jan', 'Feb', 'Mar', 'Apr']].mean().mean() for city in cities}
max_city = max(city_avg, key=city_avg.get)
max_city
```

```python
#city with the total lowest sales
cities = sales_data['city'].unique()
city_totals = {city: sales_data[sales_data['city'] == city][['Jan', 'Feb', 'Mar', 'Apr']].sum().sum() for city in cities}
min_city = min(city_totals, key=city_totals.get)
min_city
```

```python
#Add a new column Total_Sales = Jan + Feb + Mar + Apr.
sales_data['Total_Sales'] = sales_data[['Jan', 'Feb', 'Mar', 'Apr']].sum(axis=1)
```

```python
#Find the top 3 performing cities based on Total_Sales.
cities = sales_data['city'].unique()
city_totals = {city: sales_data[sales_data['city'] == city]['Total_Sales'].sum() for city in cities}
sorted(city_totals.items(), key=lambda x: x[1], reverse=True)[:3]
```

```python
# Display the top 5 stores with the highest Total_Sales.
sales_data.nlargest(5, 'Total_Sales')
```

```python
#Display the bottom 5 stores with the lowest Total_Sales.
sales_data.nsmallest(5, 'Total_Sales')
```

```python
#The store with maximum sales in March.
sales_data.loc[sales_data['Mar'].idxmax()]
```

```python
#The store with minimum sales in April.
sales_data.loc[sales_data['Apr'].idxmin()]
```

```python
#Find how many stores in each city have average monthly sales > 15.
sales_data[sales_data[['Jan','Feb','Mar','Apr']].mean(axis=1) > 15].groupby('city').size()
```

```python
#Filter stores from California that had sales > 20 in any month.
sales_data[(sales_data['city'] == 'California') & (sales_data[['Jan','Feb','Mar','Apr']] > 20).any(axis=1)]
```
