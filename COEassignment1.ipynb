```python
#Rename all columns in uppercase.
uber_data.columns=uber_data.columns.str.upper()
uber_data.columns
```

```python
#ASSIGNMENT2
#Display rides where CATEGORY == 'Business'.
business_rides=uber_data[uber_data['CATEGORY*']=='Business']
business_rides
```

```python
#Show top 5 rides with the longest distance (MILES).
top_longest_rides=uber_data.sort_values(by='MILES*',ascending=False).head(5)
top_longest_rides[['START*','STOP*','MILES*']]
```

```python
#Replace all missing PURPOSE values with "Not Specified".
uber_data['PURPOSE*'].fillna("Not Specified", inplace=True)
uber_data['PURPOSE*'].isnull().sum()

```

```python
#  Create a new column TRIP_DURATION using END_DATE - START_DATE.
uber_data['START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'], errors='coerce')
uber_data['END_DATE*'] = pd.to_datetime(uber_data['END_DATE*'], errors='coerce')
uber_data['TRIP_DURATION'] = (uber_data['END_DATE*'] - uber_data['START_DATE*']).dt.total_seconds() / 3600
uber_data[['START_DATE*', 'END_DATE*', 'TRIP_DURATION']].head()
```

```python
sorted_trips=uber_data.sort_values(by='MILES*',ascending=False)
sorted_trips[['START*','STOP*','MILES*']]
```

```python
#ASSIGNMENT3
# Group by CATEGORY* and calculate average MILES*
average_miles=uber_data.groupby('CATEGORY*')['MILES*'].mean()
average_miles
```

```python
#Find total trips for each PURPOSE.
total_trip=uber_data['PURPOSE*'].value_counts()
total_trip
```
